# Reshape Segment data

```python
def reshape_segment_data_2s(data):
    """
    This function is used for reshape and
    segment data.
    (for SVM, Random Forest, 2D-CNN, 3D-CNN)
    Parameters
    -----------
    data : tensor-like of shape (1, time_length, 64)
    
    Returns
    -------
    processed data : tensor-like of shape 
    (number_of_samples, 120, 8, 8)
    notice: 2s = 120, 1s = 60, 0.5s = 30
    """
#--------------------------------------------------------------------------------
# create two empty list
    reshape_data = []
    segment_data = []
#--------------------------------------------------------------------------------
# reshape data
# reshape 1x64 to 8x8 
    for a in range(len(data)):
        sub_sample = []
        for b in range(len(data[a])):
            sub_sample.append(np.reshape(data[a][b],(8,8)))
        reshape_data.append(sub_sample)
#--------------------------------------------------------------------------------
# segment data
    for c in range(len(reshape_data)):
        segments = list(range(0,600,120))
        for segment in segments:
            segment_data.append(reshape_data[c][segment:segment + 120])
    return segment_data
#--------------------------------------------------------------------------------
```
